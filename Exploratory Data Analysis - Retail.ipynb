```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['orange','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()

    
```

# Observation 5
<ul>
<li><B> Profit and Sales are maximum in Consumer Segment and minimum in Home office Segment</B></li>
</ul>


```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='flare')
plt.xlabel('Region',fontsize=15)
plt.show()
```

# Observation 6
<ul>
    <li><B>Segment wise sales are almost same in every region</B></li>
</ul>


```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['brown','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# Observation 7
<ul>
    <li><B>Profit and sales are higher in west region and minimum in south region</B></li>
</ul> 


```python
ps = data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['orange','blue'],figsize=(15,8))
plt.title('Profit/loss & sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

# Observation 8

<ul>
    <li><B>high profit is for california and New york</B></li>
    <li><B>Loss is for texas ,pennsylvania,ohio and llinois</B></li>
</ul>


```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# Observation 9

<ul>
    <li><B>Profit is higher in Technology and office supplies and furniture have less </B></li>
</ul>


```python
ps=data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
